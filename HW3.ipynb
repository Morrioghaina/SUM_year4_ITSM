```python
import numpy as np
import matplotlib.pyplot as plt

a, b = 2, 5
s = np.random.beta(a, b, 1000)

plt.hist(s, bins=30, density=True)
plt.title('Beta-распределение')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n, p = 10, 0.5
s = np.random.binomial(n, p, 1000)

plt.hist(s, bins=30, density=True)
plt.title('Биномиальное распределение')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

scale = 1
s = np.random.exponential(scale, 1000)

plt.hist(s, bins=30, density=True)
plt.title('Экспоненциальное распределение')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

shape, scale = 2.0, 2.0
s = np.random.gamma(shape, scale, 1000)

plt.hist(s, bins=30, density=True)
plt.title('Гамма-распределение')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

p = 0.3
s = np.random.geometric(p, 1000)

plt.hist(s, bins=30, density=True)
plt.title('Геометрическое распределение')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

mu, sigma = 0, 0.1
s = np.random.normal(mu, sigma, 1000)

plt.hist(s, bins=30, density=True)
plt.title('Нормальное распределение')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

lam = 5
s = np.random.poisson(lam, 1000)

plt.hist(s, bins=30, density=True)
plt.title('Пуассоновское распределение')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

low, high = 0, 1
s = np.random.uniform(low, high, 1000)

plt.hist(s, bins=30, density=True)
plt.title('Равномерное распределение')
plt.show()
```
